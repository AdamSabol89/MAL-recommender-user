```python
def scraper(times, delay=2):
    final_data = pd.DataFrame()
   
    for index in range(times):
        r = requests.get("https://myanimelist.net/users.php")
        time.sleep(.5)
        base = "https://myanimelist.net/animelist/"
        soup = bs(r.content)
        table = soup.find("table")
        usersdiv = table.find_all("div", attrs={"style":"margin-bottom: 7px;"})
        users = []
        
        for i in range(len(usersdiv)):
            #print(usersdiv[i].a["href"]+ ' ' + usersdiv[i].get_text())
            users.append(usersdiv[i].get_text())
        print(users)
            #print(users[0])
        for user in users:
            path = base + user
            print(path)
            time.sleep(delay)
            soup2 = bs(requests.get(path).text, 'lxml')
            #try: 
            try: 
                data = json.loads( soup2.select_one('.list-table[data-items]')['data-items'] )
            except:
                continue
            df= pd.DataFrame(data)
            df["user"] = user
            final_data = final_data.append(df)

    return final_data
            

        
```

```python
anime_data = scraper(3)
```
